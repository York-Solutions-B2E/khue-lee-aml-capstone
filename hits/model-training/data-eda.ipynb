```python
plt.figure(figsize=(5, 6), facecolor='lightblue')
sns.countplot(x='add_to_cart', data=eda_df)
plt.xlabel('Label', fontsize=10, fontweight='bold')
plt.ylabel('Total', fontsize=10, fontweight='bold')
plt.title('Target Labels', fontsize=12, fontweight='bold')
plt.xticks(fontsize=8, fontweight='bold')
plt.yticks(fontsize=8, fontweight='bold')
plt.show()
plt.show()
```

```python
filtered_subcontinent = eda_df[eda_df['sub_continent'].isin(['South America', 
                                                             'Northern America', 
                                                             'Western Europe', 
                                                             'Southeast Asia', 
                                                             'Southern Asia', 
                                                             'Northern Europe',
                                                             'Eastern Asia',
                                                             'Western Asia',
                                                             'Central America',
                                                             'Australasia'
                                                            ])]

plt.figure(figsize=(10,5), facecolor='lightblue')
sns.countplot(x = 'sub_continent', data = filtered_subcontinent, palette='inferno')
plt.xlabel('Sub-Continents', fontsize=10, fontweight='bold')
plt.ylabel('Sessions', fontsize=10, fontweight='bold')
plt.title('Sessions by Sub-Continents', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=8, fontweight='bold')
plt.yticks(fontsize=8, fontweight='bold')
plt.show()
```

```python
filtered_price = eda_df[(eda_df['product_price'] >= 0) & (eda_df['product_price'] <= 200)]

plt.figure(figsize=(6,6), facecolor='lightblue')
sns.set_style("white")
plt.gca().set_facecolor('lightyellow')
sns.scatterplot(x='add_to_cart', y='product_price', data=filtered_price, color='red')
plt.xlabel('Add to Cart Actions', fontsize=10, fontweight='bold')
plt.ylabel('Price', fontsize=10, fontweight='bold')
plt.title('Add to Cart Actions by Product Price', fontsize=12, fontweight='bold')
plt.xticks(fontsize=8, fontweight='bold')
plt.yticks(fontsize=8, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,5), facecolor='lightblue')
sns.countplot(x = 'day', data = eda_df, palette='inferno')
plt.xlabel('Day', fontsize=12, fontweight='bold')
plt.ylabel('Sessions', fontsize=12, fontweight='bold')
plt.title('Hit Sessions by Day', fontsize=14, fontweight='bold')
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.tight_layout(pad=2.0)
plt.show()
```
